# Programa para graficar las curvas de nivel
Primero hay que generar una malla de puntos, en donde evaluaremos en cada uno la función. Para ello utilizaremos `meshrid()` de NumPy:

```python
import numpy as np
xx =np.linspace(-1.5,1.5)
#tenemos una malla con 50 elementos del -1.5 al 1.5
yy = xx.copy()
#ahora la hacemos bidimensional
X,Y = np.meshgrid(xx,yy)
```

Las operaciones deben estar *vectorizadas*. Para cada fila $i$ y columna $j$ creamos el par $$(X_{ij},Y_{ij})$$. Ahora, realizaremos los siguientes pasos para evaluar.

```python
#step 1
Z = np.cos(X**2+Y**2) 
#step 2
for i in range(len(X)):
    for j in range(len(Y)):
        Z[i, j] = np.cos(X[i, j] ** 2 + Y[i, j] ** 2)
```

La matriz $Z$ tiene los valores de la función coseno para *cada punto del mallado*, es decir, es la matriz donde se almacena el valor luego de evaluar.
Ahora, tomando la función $$f(x,y)=cos(−10xy)+sinh(x+y)−log\left(\frac{3+x^2}{1+y^2}\right)$$ en el dominio $$D = [−3/2,3/2]×[−3/2,3/2]$$

```python
from numpy import cos, sinh, log
Z0 = cos(-10 * X * Y) + sinh(X + Y) - log((3 + X ** 2) / (1 + Y ** 2))
```

Representamos ahora las curvas de nivel, usando dos funciones `contour()` y `countourf()`. La primera representa solo las líneas, y en la segunda se rellena el espacio entre ellas 

```python
import matplotlib.pyplot as plt
plt.contour(X,Y,Z0)
plt.show()
cs1 = plt.contourf(X, Y, Z0, 25)  # Pintamos 25 niveles con relleno
cs2 = plt.contour(X, Y, Z0, cs1.levels, colors='k')  # Añadimos bordes negros
plt.show()
```

```python
xx_1 =np.linspace(-10,10)
yy_1 = xx_1.copy()
X1,Y1 = np.meshgrid(xx_1,yy_1)
Z1 = 2*X**2 + 4*X*Y**3-10*X*Y+Y**2
plt.contour(X1,Y1,Z1)
plt.show()
csp1 = plt.contourf(X1,Y1,Z1, 25)
css1 = plt.contour(X1, Y1, Z1, csp1.levels, colors='k')
plt.show()
```
